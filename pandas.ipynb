# 2 - How to rename Columns?

```python
mydf=pd.DataFrame({'fname':['Kiran','Kishor','Nikhil','Pawan'],
              'lname':['Pokhrel','Pokhrel','Bhusal','Pandey'],
              'age':[16,17,18,19]})
mydf
```

```python
# renaming columns
mydf.rename(columns={'fname':'first_name','lname':'last_name'},inplace=True)
mydf
```

```python
# We have to type all the columns and change there.
mydf.columns=['First_name','Last_name','age']
mydf
```

```python
# replacing chracter
mydf.columns=mydf.columns.str.replace('_','-')
mydf
```

```python
# Add prefix and suffix
mydf=mydf.add_prefix('[')
mydf=mydf.add_suffix(']')
mydf
```

```python
# Serring to defult
mydf.columns=['fname','lname','age']
mydf
```

# 11 - Splitting string into multiple columns

```python
data = {
    'name': ['John Doe', 'Alice Mary', 'Raj Kumar', 'Emily Smith', 'Michael Smith'],
    'city': ['New York', 'Los Angeles', 'Delhi', 'London', 'Chicago'],
    'phoneno': ['1234567890', '9876543210', '1122334455', '5566778899', '9988776655']
}
datadf = pd.DataFrame(data)
datadf
```

```python
datadf[["fname","lname"]]=datadf.name.str.split(" ",expand=True)
datadf
```

```python
# regining data according to need
datadf=datadf[["fname","lname","city","phoneno"]]
datadf
```
